```python
import bs4
import requests as rq
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import re
import pandas as pd
```

```python
res = rq.get('https://race.kra.co.kr/raceScore/ScoretableDetailList.do?meet=1&realRcDate=20170723&realRcNo=1')
soup = bs4.BeautifulSoup(res.text, 'html.parser')
head = soup.select('#contents > div:nth-child(9) > table > caption')
```

```python
title = []
for i in head[0].text.split(',')[1:-1]:
    title.append(i.strip())
```

```python
title.insert(0, head[0].text.split(',')[0][-2:])
title.append(head[0].text.split(',')[-1].split()[0])
```

```python
title
```

```python
data2 = {}
s = 1

for i in range(1, 12):
    res = rq.get(f'https://race.kra.co.kr/raceScore/ScoretableDetailList.do?meet=1&realRcDate=20170723&realRcNo={i}')
    soup = bs4.BeautifulSoup(res.text, 'html.parser')
    for j in range(1, 13):
        data = []
        for k in range(1, 17):
            try:
                data.append(soup.select(f'#contents > div:nth-child(9) > table > tbody > tr:nth-child({j}) > td:nth-child({k})')[0].text.strip())
            except:
                continue
                
        if len(data) == 0:
            pass
        else:
            data2[s] = data
            s += 1
```

```python
df = pd.DataFrame(data2.values(), columns=title)
```

```python
df.to_csv('./horse.csv', index=None, na_rep='')
```
