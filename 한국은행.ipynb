```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

### 코로나 데이터 전처리

```python
df = pd.read_csv('CONVENIENT_global_confirmed_cases.csv')
df = df.iloc[1:]
covid19 = df[['Country/Region', 'Korea, South']]
covid19.columns = ['년/월', '확진자 수']
covid19_v2 = covid19.copy()
covid19_v2
```

### 코로나 데이터, 년/월 column 데이터 타입 변경

```python
covid19_v2['년/월'] = covid19_v2['년/월'].apply(lambda x: pd.to_datetime(str(x)))
```

```python
covid19_v2.info()
```

### 주택매매가격지수(KB) 전처리

```python
# 주택매매가격지수(KB)
meme_price = pd.read_csv('ECOS_TABLE_20210827_153158.csv', encoding='cp949')
meme_price = meme_price.iloc[:6]
meme_price.drop(['통계표', '단위', '변환'], axis=1, inplace=True)
meme_price = meme_price.iloc[[0,2,3,4]]
meme_price = meme_price.T
meme_price = meme_price.iloc[1:].reset_index()
meme_price.columns = ['년/월', '총지수', '단독주택', '연립주택', '아파트']
meme_price
```

### 주택매매가격지수(KB) 데이터, 년/월 column 데이터 타입 변경

```python
meme_price_v2 = meme_price.copy()
```

```python
meme_price_v2['년/월'] = meme_price_v2['년/월'].apply(lambda x: pd.to_datetime(str(x)))
meme_price_v2
```

### 주택매매가격지수(KB) 데이터, 년/월을 제외한 나머지 columns 데이터 타입 변경

```python
date = meme_price_v2[['년/월']]
price = meme_price_v2[['총지수', '단독주택', '연립주택', '아파트']].astype('float')
meme_price_v2 = pd.concat([date, price], axis=1)
meme_price_v2
```

### 시각화
    * 은행 데이터 7월 까지
    * 2021년 8월 코로나 확진자 급증

```python
plt.figure(figsize=(12, 5))
plt.plot(meme_price_v2['년/월'], meme_price_v2['총지수'], label='총지수')
plt.plot(meme_price_v2['년/월'], meme_price_v2['단독주택'], label='단독주택')
plt.plot(meme_price_v2['년/월'], meme_price_v2['연립주택'], label='연립주택')
plt.plot(meme_price_v2['년/월'], meme_price_v2['아파트'],  label='아파트', c='darkblue')


plt.ylim(98, meme_price_v2['아파트'].max()*1.1)

plt.title('일별 코로나 확진자 수에 따른 KB지수 증감 여부', fontdict={'fontsize':20})
plt.xlabel('년-월')
plt.ylabel('주택가격매매지수(KB)')
plt.legend(loc='upper left')
plt.grid()

ax = plt.gca()
ax2 = ax.twinx()
ax2.plot(covid19_v2['년/월'], covid19_v2['확진자 수'], 'r--', alpha=0.8)
ax2.set_ylabel('확진자 수', color='red')
ax2.set_ylim(0, covid19_v2['확진자 수'].max()*1.1)
plt.show()
```
